```python
df = df.sort_values(['city', 'zipcode', 'date']).reset_index(drop=True)
```

```python
df
```

```python
dfsale = df.groupby(['city', 'zipcode','date'])['median_sale_price'].mean().reset_index()
```

```python
print(dfsale.head(30))
```

```python
import matplotlib.pyplot as plt
import matplotlib.ticker as mtick
import pandas as pd

# Ensure 'date' is datetime
dfsale['date'] = pd.to_datetime(dfsale['date'])

# Create 'year' column
dfsale['year'] = dfsale['date'].dt.year

# Now group by city, zipcode, year
df_yearly = dfsale.groupby(['city', 'zipcode', 'year'])['median_sale_price'].mean().reset_index()

# Plotting (optional: filter first if needed)
plt.figure(figsize=(10, 6))
plt.plot(df_yearly['year'], df_yearly['median_sale_price'])

# Format y-axis to show regular numbers
plt.gca().yaxis.set_major_formatter(mtick.StrMethodFormatter('{x:,.0f}'))

plt.title('Yearly Median Sale Price')
plt.xlabel('Year')
plt.ylabel('Median Sale Price')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import matplotlib.ticker as mtick

# Filter for one city
city_data = df_yearly[df_yearly['city'] == 'ATL']
city_data=city_data[city_data['zipcode'] == 30002]

#city_data = df_yearly[df_yearly['zipcode'] == '30002']

# Plot
plt.figure(figsize=(12, 6))
sns.lineplot(data=city_data, x='year', y='median_sale_price', hue='zipcode', legend='full')

plt.title('Yearly Median Sale Price by Zipcode (ATL)')
plt.xlabel('Year')
plt.ylabel('Median Sale Price')

# Format y-axis to show real numbers (e.g., 100000 instead of 1e5)
plt.gca().yaxis.set_major_formatter(mtick.StrMethodFormatter('{x:,.0f}'))

plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
dfsocial = df.groupby(['city', 'zipcode', 'date'])[
    ['median_sale_price','Median Home Value', 'bank', 'bus', 'hospital', 'mall', 'park',
     'restaurant', 'school', 'station', 'supermarket','Total Population']
].mean().reset_index()
```

```python
dfsocial['date'] = pd.to_datetime(dfsocial['date'])  # if not already
dfsocial['year'] = dfsocial['date'].dt.year

dfsocial_yearly = dfsocial.groupby(['city', 'zipcode', 'year'])[
    ['median_sale_price','Median Home Value', 'bank', 'bus', 'hospital', 'mall', 'park',
     'restaurant', 'school', 'station', 'supermarket','Total Population']
].mean().reset_index()

social_yearly = dfsocial_yearly[
    (dfsocial_yearly['city'] == 'ATL') & (dfsocial_yearly['zipcode'] == 30002)
]

social_features = ['median_sale_price','Median Home Value']

plt.figure(figsize=(12, 6))
for feature in social_features:
    plt.plot(social_yearly['year'], social_yearly[feature], label=feature)

plt.title('Yearly Social Infrastructure (ATL - 30002)')
plt.xlabel('Year')
plt.ylabel('Average Presence')
plt.xticks(rotation=45)
plt.legend()
plt.tight_layout()
plt.show()
```
